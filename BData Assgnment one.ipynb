```python
#### import required liabries
import pandas as pd 
```

```python
# load the dataset
loanprediction = pd.read_csv(r"C:\Users\USER\OneDrive\Desktop\loan_prediction_based_on_customer_behavior_dataset xcel.csv")
```

```python
#review the data
loanprediction.head()
```

```python
#check for missing data
loanprediction.isnull()
```

```python
#counting missing values 
loanprediction.isnull().sum()
```

```python
#dropping columns that are unnecessary for loan prediction
loanprediction= loanprediction.drop(columns=['CITY', 'STATE', 'Id','Experience'])
```

```python
#printing to show off column names
print(loanprediction.columns)
```

```python
#converting categorical data into numerical codes
loanprediction['Married/Single'] = loanprediction['Married/Single'].astype('category').cat.codes
loanprediction['House_Ownership'] = loanprediction['House_Ownership'].astype('category').cat.codes
loanprediction['Car_Ownership'] = loanprediction['Car_Ownership'].astype('category').cat.codes
loanprediction['Profession'] = loanprediction['Profession'].astype('category').cat.codes
```

```python
#review the data
loanprediction.head()
```

```python
#defining features (x) and target (y) for machine learning purposes
x = loanprediction.drop(['Risk_Flag'], axis=1).values
y = loanprediction['Risk_Flag'].values
```

```python
#importing necessary model and metrics - libraries 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

```

```python
#splitting the Data (train -test split) 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
# model Initialization (a dictionary of ML models is created for evaluation) 
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "KNN": KNeighborsClassifier(),
    "BC": BaggingClassifier()
}
```

```python
#training and evaluating models( fitting models and calculating accuracy)
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name}: Accuracy = {accuracy:.6f}")
    classification = classification_report(y_test, y_pred)
    print(f"{name}\n{classification}\n")
```
